# Kenalan dengan hacktoberfest

### Apa Itu Hacktoberfest .?

hacktoberfest merupakan sebuah ajang kontribusi untuk project open source.hacktoberfest ini disponsori oleh DigialOcean. ajang ini biasa digelar setahun sekali pada setiap bulan oktober.setiap orang bisa berkontribusi dalam ajang hacktoberfest ini.siapaun dia mau dia programmer/developer , mahasiswa atau siswa yang baru belajar pemrograman, para penyedia event(event house). maupun perusahaan manapun juga bisa berkontribusi dalam acara hacktoberfest ini .dimanapun dan siapaoun kalian asalkan kalian ingin membantu dalam perkembangan open source dan juga ingin membuat impact yang positif bagi komunitas open source. nantinya hacktoberfest menyediakan sebuah challenge. siapapun dengan latar belakang apapun bisa mengikuti dan menyelesaikan challenge ini . dan jika dapat menyelesaikan challeng bisa mendapatkan reward.Reward yang akan didapatkan nanti dapat berupa baju/ kaos dari hacktoberfest atau donasi dengan menanam pohon. kalian bisa cek lebih lanjut di website resminya di : https://hacktoberfest.digitalocean.com/



### Tujuan dari hacktoberfest

Tujuan utama dari hacktoberfest adalah mengajak kepada semua untuk ikut berkontirbusi kepada proyek proyek open source.jadi semua orang bisa berkontribusi pada proyek proyek open source. dengan banyak kontribusi pada proyek open source maka teknologi akan semakin cepat berkembang. kita bisa merasakan dimana banyak teknologi teknologi baru , framework framewrok baru yang muncul dan berkembang dengan pesat 


### Apakah kalau belum bisa coding bisa ikut berpatisipai di ajang hacktoberfest ini.?


setiap orang bisa ikut dalam berkontribusi dan berpatisipasi di acara hacktoberfest ini. tidak harus dengan menulis kode program den berkontibusi lewat kode program, kalian juga bisa berkontribusi dengan cara yang lain juga.banyak repository repository yang hanya memeliki dokumentasi dengan bahasa inggris , kalian bisa ikut berpatisipasi dengan menerjemahkan dokumentasi itu ke bahasa lokal kalian. juga bisa dengan berkontribusi dengan buku buku program yang gratis dan legal , serta bisa dengan cara lain seperti berkontribusi menulis jurnal ilmiah yang berkaitan dengan teknologi dan masih banyak lagi. 

### Setelah tau apa itu hacktoberfest , gimana cara ikutan acara ini sih.?





untuk bisa ikut berpatisipasi di hacktoberfest kamu bisa mendaftarkan diri di website hacktoberfest dan melakukan 4 kali pull request valid yang nantinya akan di cek oleh pihak hacktoberfest kevalidanya.pull request dapat dilakukan di public repo manapun, tidak harus repo yang berlabel hacktoberfest. tahun ini 70.000 participan pertaa yang sukses mengikuti challeng bakal mendapatkan rewardnya,untuk info lebih lanjut bisa di cek disini : https://hacktoberfest.digitalocean.com/details

### Daftar ke hacktoberfest

pertama tama kalian harus mempunyai akun github dan kalian harus mendaftar website hacktoberfest di https://hacktoberfest.digitalocean.com/.
lalu bisa lanjut ke start hacking.



jika kalian belum pernah mendaftar maka bisa mendaftarkan diri terlebih dahulu dengan melakukan sign up dengan akun github dan juga nantinya diminta untuk mengisi akun email yang sesuai dengan akun github.

lalu kalian sudah mendaftar bisa masuk/sign in lewat akun github 


setelah kalian berhasil masuk dan mendaftarkan akun github kalian di website hacktoberfest kalian sekarang bisa berkontirbusi dengan melakukan pull request.kalian bisa melakukan minimal 4 pull request lebih juga gapapa. progress pull request kalian bisa dilihat di halaman profil websitenya



nah juga perlu diingat hacktoberfest ini hanya berlaku dibulan oktober dari tanggal 1 sampai 31 oktober.jadi diluar bulan oktober pull request tidak dihitung.nantinya juga pull request yang kalian lakukan akan diperiksa terlebih dahulu dari pihak hacktoberfest apakah valid atau tidak. so , jangan pull request asal asalan dan lakukan pull request yang valid ya.. :)

### Cara melakukan pull request

untuk melakukan pull request ada beberapa cara dan langkah langkahnya. diantaranya bisa melakukan lewat gui dan juga bosa dilakukan melalui terminal / command line.

###### Dan perlu diingat sebelum kalian ingin melakukan kontribusi kalian harus membaca peraturan kontribusi pada sebuah projek di github yang ingin kalian berkontribusi di dalamnya

untuk melihat aturan kontribusi kalian bisa melihat di repo github yang kalian ingin berkontribusi didalamnya. biasanya tertulis di file READ.md ataupun CONTRIBUTING.md . karena aturan dan tatacara berkontibusi harus kita patuhi bersama juga.jadi jangan lupa untuk meminta izin serta membaca aturan aturan tersebut dengan teliti dan seksama.

#### 1. fork dan clone Repository
 untuk melakukan fork dan clone repo kalian harus terlebih dahulu masuk ke akun github. setelah itu kalian bisa melakukan fork dan clone. fork sendiri bisa kita artikan menyalin repo dari akun lain atau organisasi lain ke repo kita
